```python
import random
from random import randint
import string
import pandas as pd 
import csv

# Global variables
char_string = string.ascii_letters 
char_digits = string.digits

# DataFrame creation 
columns = ["Website", "Password"]
df = pd.DataFrame(columns=columns)

# Function definitions

def easy_password():
    ''' generating an easy password containing a string with 4 letters and 4 digits. '''
    
    easy_char = ''.join(random.choice(char_string) for i in range(4))
    
    easy_digits = ''.join(random.choice(char_digits) for i in range(4))

    pw_easy = easy_char + easy_digits
    easy = ''.join(random.sample(pw_easy,len(pw_easy)))
    return easy


def medium_password():
    ''' generating a medium password containing a string with 5 letters and 5 digits with 
        atleast two upper- and lowercase. 
    '''
    
    medium_char_low = ''.join(random.choice(char_string).lower() for i in range(3))
    medium_char_upp = ''.join(random.choice(char_string).upper() for i in range(2))
    
    medium_digits = ''.join(random.choice(char_digits) for i in range(5))
    pw_medium = medium_char_low + medium_char_upp + medium_digits
    medium = ''.join(random.sample(pw_medium,len(pw_medium)))
    return medium

def strong_password():
    ''' generating a medium password containing a string with 5 letters, 5 digits and 
        2 special charachters with atleast two upper- and lowercase. 
    '''
    
    strong_char_low = ''.join(random.choice(char_string).lower() for i in range(3))
    strong_char_upp = ''.join(random.choice(char_string).upper() for i in range(3))
    
    strong_digits = ''.join(random.choice(char_digits) for i in range(5))
    
    spec_char = [i for i in string.punctuation]
    first_spec_char = random.choice(string.punctuation)
    second_spec_char = random.choice(spec_char)
    spec_charafirst = first_spec_char + second_spec_char
     
    pw_strong = strong_char_low + strong_char_upp + strong_digits + spec_charafirst
    strong = ''.join(random.sample(pw_strong,len(pw_strong)))
    return strong

def auto_password():
    ''' generating a medium password containing a string with 5 letters, 5 digits and 
        2 special charachters with atleast two upper- and lowercase. 
    '''
    while True:
        try:
            password_upper = int(input("How many uppercase letter do you want?: "))
            password_lower = int(input("How many lowercase letter do you want?: "))
            password_digit = int(input("How many digits letter do you want?: "))
            password_spec = int(input("How many special characters do you want?: "))
            
            minimum_length = password_upper + password_lower + password_digit + password_spec

            if minimum_length < 8:
                print('The minimum requirement length must be greater than 8 letters!')
            else:
                lower = ''.join(random.choice(char_string).lower() for i in range(password_lower))
                upper = ''.join(random.choice(char_string).upper() for i in range( password_upper))
                digits = ''.join(random.choice(char_digits) for i in range(password_digit))

                spec_char = [i for i in string.punctuation]
                lst = []
                for i in range(password_spec):
                     lst.append(random.choice(spec_char))

                string_of_spec_char = "".join(lst)

                custom_password = lower + upper + digits + string_of_spec_char

                random_custom = ''.join(random.sample(custom_password,len(custom_password)))
                return random_custom
        except ValueError:
            print('Enter only numbers.')



def user_choice():
    ''' this function displays what level of password the user wants. '''
    
    print ("Easy password contains four digits and four letters with a total length of 8.\n")
    print("Medium password contains five digits and five letters with atleast 2 upper and lowercase with total length of 10.\n")
    print("Strong password contains five digits and five letters with atleast 2 upper and lowercase a total and two special characterwith length of 10.\n")


    print('Choose what kind of difficulty you want for your password. Choose between 1-4: \n') 
    choice = ''
    
    
    while True: 
        choice = input(' 1. Easy\n 2. Medium \n 3. Strong\n')
        if choice == '1':
            result = easy_password()
            return result
        
        elif choice == '2':
            result = medium_password()
            return result
        
        elif choice == '3':
            result = strong_password()
            return result
        
        else:
            print('Choose between 1-4!')
            

def save_password(choice):
    ''' this function takes the saved password and asks the user what the of the password 
        will be and then saves add the name and the password to an existing CSV-file. 
    '''


    df.append(pd.Series(), ignore_index=True)
    df.loc[df.shape[0], "Website"] = input("What website is the password for? ")
    #save1= df.loc[df.shape[0], "Website"]
    df.loc[df.shape[0]-1, "Password"] = choice
    #df.index = np.arange(1, len(df) + 1)
    df.to_csv('passwords.csv', index=False)
    
    return df


def reading_DF():
    ''' function that reads the csv file. '''
    file = 'passwords.csv'
    df = pd.read_csv(file)
    return df
    


def feedback(choice):
    ''' this function clears the csv file  '''
    
    
    print('Do you want to clear csv-file? Choose between 1-2: \n') 
    
    save = ''
    while True:
        save = input(' 1. Yes\n 2. No\n\n')
        if save == '1':
            f = open('passwords.csv', "w")
            f.truncate()
            f.close()
            break
        else:
            print('\nGoodbye')
            break       


print("Welcome to Password Generation software\n")

while True:
    
    menu = ''
    menu = input(' 1. Generate a password\n 2. Display password\n 3. Clear password\n 4. Customize password\n 5. EXIT\n\n ')
    
    if menu == '1':
        password = user_choice()
        choice = save_password(password)        
        
    elif menu == '2':
        with open('passwords.csv', 'r') as csvfile:
            csv_dict = [row for row in csv.DictReader(csvfile)]
            if len(csv_dict) == 0:
                print('The csv file is empty!')
            else:
                table = reading_DF()
                print(table)
            

    elif menu == '3':
        feedback(choice)
    
    elif menu == '4':
        print('\nMinimum 8 characters in total!')
        password = auto_password()
        choice = save_password(password)
    elif menu == '5':
        print('Goodbye!')
        break
    else:
        print('Choose between 1-4!')
        

        
        
    
```
